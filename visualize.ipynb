```python
import pandas as pd
import numpy as np

df = pd.read_csv('./ms_data.csv')
df['visit_date'] = pd.to_datetime(df['visit_date'])
df['patient_id'] = df['patient_id'].astype(str)
df['age'] = df['age'].astype(float)
df['education_level'] = df['education_level'].astype('category')
df['walking_speed'] = df['walking_speed'].astype(float)
# sort by patient_id and visit_date
df = df.sort_values(by=['patient_id', 'visit_date'])
df.dropna()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
sns.lmplot(x='age', y='walking_speed', data=df, aspect=1.5)
plt.title('Scatter plot of Age vs Walking Speed with Regression Line')
plt.xlabel('Age')
plt.ylabel('Walking Speed')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='education_level', y='walking_speed', data=df)
plt.title('Box Plot of Walking Speed by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Walking Speed')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='age', y='walking_speed', hue='education_level', data=df)
plt.title('Line Plot of Walking Speed by Age and Education Level')
plt.xlabel('Age')
plt.ylabel('Walking Speed')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='insurance_type', y='visit_cost', data=df, ci='sd')
plt.title('Bar Plot of Mean Visit Costs by Insurance Type')
plt.xlabel('Insurance Type')
plt.ylabel('Mean Visit Cost')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='insurance_type', y='visit_cost', data=df)
plt.title('Box Plot of Visit Costs by Insurance Type')
plt.xlabel('Insurance Type')
plt.ylabel('Visit Cost')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.pairplot(df[['age', 'walking_speed', 'visit_cost', 'education_level', 'insurance_type']], hue='education_level')
plt.title('Pair Plot of Key Variables')
plt.show()

```

```python
g = sns.FacetGrid(df, col='education_level', row='insurance_type', margin_titles=True)
g.map(sns.scatterplot, 'age', 'walking_speed')
g.add_legend()
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='visit_date', y='walking_speed', hue='education_level', data=df)
plt.title('Time Trend of Walking Speed by Education Level')
plt.xlabel('Visit Date')
plt.ylabel('Walking Speed')
plt.show()

```

```python
plt.figure(figsize=(10, 6))
sns.lineplot(x='visit_date', y='visit_cost', hue='insurance_type', data=df)
plt.title('Time Trend of Visit Cost by Insurance Type')
plt.xlabel('Visit Date')
plt.ylabel('Visit Cost')
plt.show()
```
